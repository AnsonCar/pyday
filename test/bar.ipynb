```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rc("font",family='Heiti TC')
```

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
plt.bar(fruits,counts)
plt.show()
```

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
plt.barh(fruits,counts)
plt.show()
```
